# API

```python
# last available day for solar radiation "20221101"
start = "20221101"
end = "20221101"
location = "latitude=39.46&longitude=-0.3"
parameters = "ALLSKY_SFC_LW_DWN;ALLSKY_KT"
community = "re" # "ag" "sb" "re"
url = f"https://power.larc.nasa.gov/api/temporal/hourly/point?start={start}&end={end}&{location}&community={community}&parameters={parameters}&format=json&user=cam&header=true&time-standard=utc"

```

```python
response = requests.get(url)
json.loads(response.content)
```

```python
# request data availability 
url = "https://power.larc.nasa.gov/api/system/resources/dashboard/Availability"
response = requests.get(url)
json.loads(response.content)
```

# Example from docs

```python
import fsspec

import xarray as xr

filepath_month = 'https://power-analysis-ready-datastore.s3.amazonaws.com/power_901_monthly_radiation_utc.zarr'
filepath_hour = 'https://power-analysis-ready-datastore.s3.amazonaws.com/power_901_hourly_radiation_utc.zarr'

filepath_mapped_month = fsspec.get_mapper(filepath_month)
ds = xr.open_zarr(store=filepath_mapped_month, consolidated=True)

filepath_mapped_hour = fsspec.get_mapper(filepath_hour)
ds_hour = xr.open_zarr(store=filepath_mapped_hour, consolidated=True)
```

```python
# accesing values of one dimension
ds["time"]
ds["lat"]
ds["lon"];
```

```python
# accesing specific measurement
ds.AIRMASS # all values in the three coordinates
```

```python
# accesing a variable in a specified range
ds["AIRMASS"].sel(time="2001-01")
```
