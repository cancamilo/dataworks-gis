```python
df_flat = df.reset_index()
df_flat.head(2)
```

```python
df_flat["geo_id"] = df_flat.index
df_flat;
```

```python
df_flat["CLRSKY_DAYS"] = df_flat["CLRSKY_DAYS"].astype(str)

# Point(lat,long)
df_flat["COORDS"] = df_flat.apply(lambda row: Point(row["lat"], row["lon"]).wkb, axis = 1)
```

```python
#determine schema
type_dict = {
    'b' : 'BOOLEAN',
    'i' : 'INTEGER',
    'f' : 'FLOAT',
    'O' : 'STRING',
    'S' : 'STRING',
    'U' : 'STRING',
    'M': 'TIMESTAMP'
}

schema = [
    {
        'name' : col_name, 
        'type' : "GEOGRAPHY" if col_name == "COORDS" else type_dict.get(col_type.kind, 'STRING')
    } for (col_name, col_type) in df_flat.dtypes.iteritems()
]

df_json = pd.DataFrame({
    col: (df_flat[col] if col != 'COORDS' else df_flat[col].map(lambda x: json.dumps(shapely.geometry.mapping(make_valid(x)))))
    for col in df_flat
})

# for (col_name, col_type) in df_flat.dtypes.iteritems():
#     print("================")
#     print(col_name)
#     print(col_type)
#     print(col_type.kind)

schema
```
